# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data = school_data_complete[['School ID','school_name', 'type', 'size', 'budget']].drop_duplicates()

TotalSchools = len(school_data)
TotalStudents = len(school_data_complete)
TotalBudget = school_data['budget'].sum()
AverageMath = school_data_complete['math_score'].mean()
AverageReading = school_data_complete['reading_score'].mean()

PassingMathDF = school_data_complete.loc[school_data_complete['math_score'] >= 70, :]
PassingReadingDF = school_data_complete.loc[school_data_complete['reading_score'] >= 70, :]

PassingMath = len(PassingMathDF)/TotalStudents*100
PassingReading = len(PassingReadingDF)/TotalStudents*100
PassingBoth = (AverageMath + AverageReading)/2

DistrictSummary = pd.DataFrame({"Total Schools":[TotalSchools],
                                "Total Students":[TotalStudents],
                                "Total Budget":[TotalBudget],
                                "Average Math Score":[AverageMath],
                                "Average Reading Score":[AverageReading],
                                "% Passing Math":[PassingMath],
                                "% Passing Reading":[PassingReading],
                                "% Overall Passing Rate":[PassingBoth]
                               })
DistrictSummary["Total Budget"] = DistrictSummary["Total Budget"].map("${:,.2f}".format)
DistrictSummary["Total Students"] = DistrictSummary["Total Students"].map("{:,}".format)
DistrictSummary["% Passing Math"] = DistrictSummary["% Passing Math"].map("%{:,.5f}".format)
DistrictSummary["% Passing Reading"] = DistrictSummary["% Passing Reading"].map("%{:,.5f}".format)
DistrictSummary["% Overall Passing Rate"] = DistrictSummary["% Overall Passing Rate"].map("%{:,.5f}".format)

DistrictSummary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python

school_data_grouped = school_data_complete.groupby(["school_name"])
school_type = school_data_grouped["type"].unique().str[0]
school_list = school_data_grouped["school_name"].unique().str[0]
school_students = school_data_grouped["Student ID"].count()
school_budget = school_data_grouped["budget"].unique().str[0]

student_budget = school_budget/school_students
    
reading_average = school_data_grouped["reading_score"].mean()
math_average = school_data_grouped["math_score"].mean()    

reading_data = school_data_complete.loc[school_data_complete['reading_score'] >= 70, :].groupby(['school_name'])
reading_pass = reading_data['school_name'].count()/school_students*100

math_data = school_data_complete.loc[school_data_complete['math_score'] >= 70, :].groupby(['school_name'])
math_pass = math_data['school_name'].count()/school_students*100

total_pass = (reading_pass+math_pass)/2

s = pd.DataFrame(list(zip(school_list, school_type, school_students, school_budget, student_budget,
                         reading_average, math_average, reading_pass, math_pass, total_pass)),
                columns=["","School Type","Total Students","Total School Budget", "Per Student Budget",
                         "Average Reading Score","Average Math Score", "% Passing Reading",
                         "% Passing Math","% Overall Passing Rate"]).set_index("")
s["Total School Budget"] = s["Total School Budget"].map("${:,.2f}".format)
s["Per Student Budget"] = s["Per Student Budget"].map("${:,.2f}".format)

s.sort_values(by='% Overall Passing Rate', ascending=False, inplace=True)
s.head(5)
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
s.sort_values(by='% Overall Passing Rate', ascending=True, inplace=True)
s.head(5)
```
